```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from tqdm import tqdm
```

```python
number_of_pages=int(input())
page_list=[]
for i in tqdm(range(1,(int(number_of_pages))+1)):
    url='https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+str(i)
    page_list.append(url)
    print(url)
    
```

```python
mobile_url=[]
mobile_name=[]
url_list=[]
mobile_price=[]
mobile_strike_Price=[]
mobile_rating=[]
mobile_desc=[]
```

```python
for page in tqdm(page_list):
    page_to_scrape = requests.get(page)
    soup = BeautifulSoup(page_to_scrape.text, "html.parser")
    box = soup.find_all("div", {'class':"_2kHMtA"})
    
    for mobile_section in box:
        name = mobile_section.find('div', class_="_4rR01T").text
        link = mobile_section.find("a", class_="_1fQZEK").get("href")
        price = mobile_section.find("div", class_="_30jeq3 _1_WHN1").text
        rating = mobile_section.find("div", class_="_3LWZlK").text
        Strike_price=mobile_section.find("div",class_="_3I9_wc _27UcVY")
        descriptions = [li.text for li in mobile_section.find_all("li", class_="rgWa7D")]


        # Append the extracted information to respective lists
        mobile_url.append(link)        
        mobile_name.append(name)
        mobile_price.append(price)
        mobile_rating.append(rating)
        mobile_strike_Price.append(Strike_price.text if Strike_price else None)
        mobile_desc.append(descriptions)
```

```python
domain_name="https://filpkart.com"
for i in mobile_url:
    url=domain_name+i
    url_list.append(url)
    print(url_list)
```

```python
dict_mobile={
    'mobile_url':url_list,
    'mobile_name':mobile_name,
    'mobile_price':mobile_price,
    'mobile_rating':mobile_rating,
    'mobile_strike_Price':mobile_strike_Price,
    'mobile_desc':mobile_desc
}
```
